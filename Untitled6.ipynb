*Outliers*** 

```python
def DetecciondeOutliers(df): #creo una funcion para detectar outliers en las columnas numericas 

    avg, std = np.mean(df), np.std(df)
    #calculo un corte donde considerare que a partir de alli es un minimo o maximo
    corte = std * 3
    #establezco minimos y maximos
    min, max = avg - corte, avg + corte
    #creo 3 listas una contendra los outliers que estan por fuera del minimo establecido, del maximo y la cantidad de outliers en total 
    outliers_min = [x for x in df if x < min]
    outliers_max = [x for x in df if x > max]
    outliers_total = [x for x in df if x < min or x > max]

    # creo un array sin outliers
    sinoutliers = [x for x in df if x > min and x < max]
    
    print('outliers identificados por debajo del minimo: %d' % len(outliers_min)) 
    print('outliers identificados por encima del maximo: %d' % len(outliers_max)) 
    print('Total de outliers identificados: %d' % len(outliers_total)) 
    print('No poseen outlier: %d' % len(sinoutliers)) 
    print("porcentaje total de outliers: ", round((len(outliers_total) / len(sinoutliers) )*100, 4)) #aqui establezco el total de outliers y le digo que quiero visualizarlo con 4 decimales
    
    return
```

**Modelo de prediccion que dira si el usuario realizara o no una compra en nuestro sitio**

```python
df['year'] = df['date'].dt.year #para preparar la data vamos a separar la fecha en diversas columnas puesto que será util al momento de aplicar el modelo
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df = df.drop(['date','fullVisitorID'], axis=1) #borramos las columnas que no vamos a precisar
```

```python
dummies=['source','deviceCategory','country','city','channelGrouping','browser'] #creamos una lista con las variables categoricas 
```

```python
dfdummies= pd.get_dummies(df[dummies], drop_first=True) #transformamos las variables categorias contenidas en la listaen numericas a traves del metodo dummie que crea una columna por cada variable existente dentro de la columna y si esta pone 1 y sino 0
```

```python
df = pd.concat([df, dfdummies],axis=1)#agrego al dataframe las columnas de las variables dummies creadas
df.drop(dummies, axis=1, inplace=True)#borro del dataframe las columnas contenidas dentro de dummies, que son las categoricas sin tratar
df.shape
```

```python
X = df.drop('transactions',axis=1) #determino mi eje x e Y
y = df['transactions']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42) #separo los datos en entrenamiento y prueba dejando un 70%, 30%
X_train.shape, X_test.shape
```

```python
model = DecisionTreeClassifier(random_state=42, max_depth=4) #llamo al modelo a utilizar en este caso un arbol de decision y determino los parametros a utilizar
```

```python
model.fit(X_train, y_train)#entreno el modelo con los datos
```

```python
y_train_pred = model.predict(X_train)#creo los datos que debera predecir
y_test_pred = model.predict(X_test)
```

```python
print ('Precision: ', accuracy_score(y_test, y_test_pred)) #se observa que el modelo tiene un 0.99 de precision
print ('\n Informacion de clasificación:\n', classification_report(y_test,y_test_pred)) #vemos el % total de predicciones + correctas respecto al total de predicciones positivas y en relacion al total de positivos reales, pondera la presicion y recuperacion y da un total de casos analizados 
print ('\n Matriz de confusion:\n',confusion_matrix(y_test, y_test_pred))
```
